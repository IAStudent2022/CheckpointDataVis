```python
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map(df)
plt.show()
```

```python
#Interprétation de la map
#The correlation matrix shows the correlation values, which measure the degree of linear relationship 
#between each pair of variables. 
#a color code between dark blue (to say that the attributes are very poorly correlated) 
#and dark red (to say that the attributes are strongly correlated).
#Moreover, on the diagonal we find the dark red corresponding to a correlation equal to 1 between 2 identical attributes.
#Here we notice that the attributes SibSp and Parch are the most correlated
#For the others the correlation is weak

```
